```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
##from https://stackoverflow.com/questions/23246125/how-to-center-labels-in-histogram-plot
def bins_labels(bins, **kwargs):
    bin_w = (max(bins) - min(bins)) / (len(bins) - 1)
    plt.xticks(np.arange(min(bins)+bin_w/2.0, max(bins), bin_w), bins, **kwargs)
    plt.xlim(bins[0], bins[-1])
```

## Edge distribution (transcript)

```python
df_txome = pd.read_table("/home/hirak/Projects/pufferfish/build/edge.txome.table",sep="\t",
                   header=None,names=['Node','Degree'])
```

```python
bins = range(10)
plt.hist(df_txome.Degree, bins=bins)
bins_labels(bins, fontsize=10)
plt.yscale('log')
#plt.xlim(0,8)
plt.title("Human Transcriptome")
plt.ylabel('Frequency (log scale)')
plt.xlabel('Degree')
plt.show()
```

```python
sns.distplot(df_txome.Degree,bins=8, norm_hist=False)
```
